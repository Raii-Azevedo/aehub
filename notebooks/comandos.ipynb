# 🛠️ Django Essential Commands Guide

Quick reference for Django development, deployment, and terminal commands.

````python
# Windows
venv\Scripts\activate

# Linux/Mac
source venv/bin/activate
````

````python
# Instalar tudo do requirements.txt
pip install -r requirements.txt

# OU instalar pacotes individuais:
pip install django
pip install psycopg2-binary
pip install python-dotenv
pip install django-htmx django-bootstrap5
pip install pillow
````

````python
# ===== CRIAÇÃO E CONFIGURAÇÃO =====

# Criar novo projeto
django-admin startproject config .

# Criar novo app
python manage.py startapp core

# ===== MIGRAÇÕES =====

# Criar migrações
python manage.py makemigrations

# Aplicar migrações (criar tabelas no banco)
python manage.py migrate

# Ver SQL de uma migração específica
python manage.py sqlmigrate core 0001

# ===== USUÁRIO E ADMIN =====

# Criar superuser (admin)
python manage.py createsuperuser

# ===== SERVIDOR =====

# Rodar servidor de desenvolvimento (padrão: localhost:8000)
python manage.py runserver

# Rodar servidor em porta diferente
python manage.py runserver 8080

# ===== UTILITÁRIOS =====

# Abrir shell Django (para testar código)
python manage.py shell

# Coletar arquivos estáticos
python manage.py collectstatic

# Verificar problemas no projeto
python manage.py check
````

````python
# Ver status das alterações
git status

# Adicionar arquivos para staging
git add .                    # Adiciona tudo
git add core/models.py       # Adiciona arquivo específico

# Fazer commit com mensagem
git commit -m "mensagem do commit"

# Enviar alterações para GitHub
git push

# Puxar atualizações do repositório remoto
git pull
````

````python
# Login na Railway
railway login

# Conectar projeto local com projeto no Railway
railway link

# Criar túnel para acessar banco de dados remoto localmente
railway connect

# Deploy (fazer upload das mudanças)
railway up

# Ver logs da aplicação
railway logs
````

````python
# ===== NAVEGAÇÃO =====

# Navegar entre pastas
cd nome_da_pasta

# Voltar uma pasta
cd ..

# ===== LISTAR ARQUIVOS =====

# Ver conteúdo da pasta (Windows)
dir

# Ver conteúdo da pasta com detalhes (Linux/Mac)
ls -la

# Ver arquivos ocultos (Windows)
dir /a

# ===== CRIAR ARQUIVOS E PASTAS =====

# Criar pasta
mkdir nome_da_pasta

# Criar arquivo vazio (Windows)
type nul > arquivo.txt

# Criar arquivo vazio (Linux/Mac)
touch arquivo.txt

# ===== LIMPAR TELA =====

# Limpar tela (Windows)
cls

# Limpar tela (Linux/Mac)
clear
````

## 💡 Dicas Importantes

### Verificar Ativação do VENV
Você saberá que o VENV está ativado quando vir `(venv)` no início da linha do terminal:
```
(venv) C:\Users\Raiis\Desktop\ae_knowledge_hub>
```

### Ordem de Execução
1. **Sempre** ative o VENV em primeiro
2. Crie/atualize migrações com `makemigrations`
3. Aplique as migrações com `migrate`
4. Crie o superuser apenas **na primeira vez**
5. Inicie o servidor com `runserver`

### Parar o Servidor
Pressione `Ctrl + C` no terminal para parar o servidor de desenvolvimento.

### Acessar Admin
- URL: `http://localhost:8000/admin`
- Use as credenciais do superuser que criou

### Banco de Dados
- Por padrão, Django usa SQLite (`db.sqlite3`)
- Para produção, use PostgreSQL configurado no Railway
